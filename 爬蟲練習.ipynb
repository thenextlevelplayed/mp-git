```python
title_data = []
Product_Type_data = []
price_excl_tax_data = []
price_incl_tax_data = []
description_data = []
for i in href_data:
    res2 = requests.get(i)
    soup = BeautifulSoup(res2.text)
    
    title_data.append(soup.select('h1')[0].text)
    
#     soup.select('.table td')[1]
    Product_Type_data.append(soup.select('.table td')[1].text)
    

    price_excl_tax_data.append(soup.select('.table td')[2].text)
    price_incl_tax_data.append(soup.select('.table td')[3].text)

    
#     soup.select('article > p')
    description_data.append(soup.select('article > p')[0].text)
    
    
price_excl_tax_data = [w.replace('Â','') for w in price_excl_tax_data]
price_incl_tax_data = [w.replace('Â','') for w in price_incl_tax_data]
    
```

```python
from pandas import DataFrame
df = pd.DataFrame(list(zip(title_data, Product_Type_data, price_excl_tax_data, price_incl_tax_data, description_data)),
               columns =['title', 'Product_Type', 'price_excl_tax', 'price_incl_tax', 'description'])
```

時間迴圈


```python
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2021, 1, 16)
end_date = date(2021, 2, 1)
ntpc_data_range = []
for single_date in daterange(start_date, end_date):
    ntpc_data_range.append(single_date.strftime("%Y-%m-%d"))
```

```python
ntpc_data_range
```

```python
a= [j for i in ntpc_station for j in ntpc_data_range]
# for i in ntpc_station:
#     for j in ntpc_data_range:
#         print(j)
```

```python
import random
import time
from random import randint
from time import sleep

ntpc_rainfall_station = pd.DataFrame()


for i in ntpc_station_time:
    time.sleep(randint(10,60))
    s=i[1]
    payload = {
        "command": "viewMain",
        "station": i[0],
        "stname": requests.utils.quote(s.encode('utf8')),
        "datepicker": i[2],
        }
    res = requests.get('https://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do', params=payload)
    test = pd.read_html(res.text)


    for index in range(1, len(test),1):  #從0開始到表格個數，遞增值為1
        test_df = test[index] ## 轉置表格
        test_df['ID'] = str(test[0][1]).split('\n')[0]
        test_df['updatetime'] = i[2]
        ntpc_rainfall_station = pd.concat([ntpc_rainfall_station, test_df], ignore_index=True)

# 因爬的次數過於頻繁被封鎖IP
```
